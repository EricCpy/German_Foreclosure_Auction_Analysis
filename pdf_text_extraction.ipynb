# PyPDF2
- liefert unformatierten PDF text, beachtet Tabellen etc nicht

```python
from pypdf import PdfReader

def pdf_to_string(file_path):
    with open(file_path, "rb") as file:
        pdf_reader = PdfReader(file)
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text() or ""
    return text

pdf_text = pdf_to_string(file_path)
print(pdf_text)
```

# PyMuPDF
- formatiert teilweise besser als pypdf und bietet mehr Funktionen, ich brauche aber nur den text wäre wahrscheinlich trotzdem sinnvoll pymupdf dafür zu verwenden, durch die bessere formatierung
- kann Tabellen etc beachten

```python
import pymupdf

def pdf_to_string(file_path):
    pdf_document = pymupdf.open(file_path)
    text = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document[page_num]
        text += page.get_text()
    pdf_document.close()
    return text

pdf_text = pdf_to_string(file_path)
print(pdf_text)
```
